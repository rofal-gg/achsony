# 1. SOAL 1 (Bobot: 45%)
Buatlah sebuah program untuk menambahkan **satu** atau **beberapa** mahasiswa dengan
kriteria sebagai berikut:
- Program meminta user, berapa data mahasiswa yang akan dimasukkan
- Program meminta data mahasiswa yang akan disimpan ke dalam dictionary
dengan struktur:
- **Key** : menggunakan NIM singkat (Contoh: 24-009)
- **Value** : dengan format
  - Nama (string)
  - Prodi (string)
  - Angkatan (integer)
  - IPK (float)

- Tambahkan data-data yang telah dimasukkan oleh user ke dalam dictionary.
- Tampilkan mahasiswa yang telah ditambahkan kedalam dictionary dengan
menggunakan looping.

**Contoh Output:**

image.png

```python
data=int(input("berapa mahasiswa ingin ditambahkan?: "))
print()
data_mahasiswa={}
for i in range (data):
  nim=(input("Masukkan NIM: "))
  nama=(input("Masukkan Nama: "))
  prodi=(input("Masukkkan Prodi: "))
  angkatan=int(input("masukkan Angkatan: "))
  ipk=float(input("masukkkan IPK: "))
  data_mahasiswa[nim]=[nama, prodi, angkatan, ipk]
  print()
print("======================")
print()
for nim, value in data_mahasiswa.items():
  print(f"NIM: {nim}")
  print(f"nama: {value[0]}")
  print(f"prodi: {value[1]}")
  print(f"angkatan: {value[2]}")
  print(f"ipk: {value[3]}")
  print()
print("======================")
```

# 2. SOAL 2 (Bobot: 55%)
Seorang dosen ingin menganalisis nilai praktikum mahasiswa. Buatlah program yang
mengelola dan memfilter nilai-nilai ini.

**Nb: Dilarang menggunakan fungsi/sintaks selain in, append(), len(), dan insert() untuk memanipulasi list.**
- Buatlah sebuah **list** bernama **daftar_nilai** yang berisi sepuluh (10) angka bulat
acak, yaitu **[80, 90, 65, 34, 40, 50, 70, 85, 44, 20]**.
- Buatlah dua buah **list baru** bernama **nilai_lulus** dan **nilai_gagal**.
- Gunakan **konsep perulangan** untuk mengiterasi seluruh elemen di
**daftar_nilai**.
  - Jika nilai >= 70, masukkan nilai tersebut ke dalam **nilai_lulus**.
  - Jika nilai < 70, masukkan nilai tersebut ke dalam **nilai_gagal**.
- Terdapat beberapa output yang ditampilkan, yaitu:
  - List dari **daftar_nilai**, **nilai_gagal**, dan **nilai_lulus**
  - Nilai rata-rata dari masing-masing **daftar_nilai**, **nilai_gagal**, dan
  **nilai_lulus**.
  - Total **jumlah** mahasiswa yang **lulus** & **gagal**.
  (Hint: Gunakan method **len()** untuk menghitung panjang dari tipe data
  koleksi)

**Contoh Output:**

image.png

```python
daftar_nilai = [80, 90, 65, 34, 40, 50, 70, 85, 44, 20]
nilai_lulus = []
nilai_gagal = []
for nilai in daftar_nilai:
    if nilai >= 70:
        nilai_lulus.append(nilai)
    else:
        nilai_gagal.append(nilai)

print("Daftar nilai:", daftar_nilai)
print("Nilai lulus:", nilai_lulus)
print("Nilai gagal:", nilai_gagal)
print()

total_nilai = 0
for n in daftar_nilai:
    total_nilai += n
rata_nilai = total_nilai / len(daftar_nilai)

total_lulus = 0
for n in nilai_lulus:
    total_lulus += n
rata_lulus = total_lulus / len(nilai_lulus)

total_gagal = 0
for n in nilai_gagal:
    total_gagal += n
rata_gagal = total_gagal / len(nilai_gagal)

print("Nilai rata-rata daftar nilai:", rata_nilai)
print("Nilai rata-rata lulus:", rata_lulus)
print("Nilai rata-rata gagal:", rata_gagal)
print()
print("Jumlah mahasiswa yang lulus:", len(nilai_lulus))
print("Jumlah mahasiswa yang gagal:", len(nilai_gagal))
```
